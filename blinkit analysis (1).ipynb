### **KPI's requirements**

```python
# total sales
total_sales = df['Sales'].sum()

# Average sales
avg_sales = df['Sales'].mean()

# no of item
no_of_item_sold = df['Sales'].count()

# Average ratings
avg_ratings = df['Rating'].mean()

# display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.1f}")
print(f"No Of Item Sold: {no_of_item_sold:,.0f}")
print(f"Average Ratings: {avg_ratings:.1f}")
```

### **chart requirements**

### **Total sales by fat content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle=90
)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

### **Total sales by item type**

```python
import matplotlib.pyplot as plt

# Total sales by item type
sales_by_item = df.groupby('Item Type')['Sales'].sum()

# Bar chart
plt.figure(figsize=(10,6))

plt.bar(sales_by_item.index, sales_by_item.values)

plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

plt.xticks(rotation=45)

plt.show()
```

### **fat content level by outlet for total sales**

```python
import matplotlib.pyplot as plt

# Group sales by outlet and fat content
sales_fat_outlet = df.groupby(
    ['Outlet Type', 'Item Fat Content']
)['Sales'].sum().unstack()

# Plot
sales_fat_outlet.plot(
    kind='bar',
    figsize=(10,6)
)

plt.xlabel('Outlet Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Fat Content and Outlet Type')

plt.xticks(rotation=45)
plt.legend(title='Fat Content')

plt.show()
```

### **total sales by outlet establishment**

```python
import matplotlib.pyplot as plt

# Total sales by outlet establishment year
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum()

# Plot
plt.figure(figsize=(10,6))

plt.plot(
    sales_by_year.index,
    sales_by_year.values,
    marker='o'
)

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')

plt.xticks(rotation=45)

plt.show()
```

### **sales by outlet location type**

```python
import matplotlib.pyplot as plt

# Total sales by outlet location type
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum()

# Bar chart
plt.figure(figsize=(8,5))

plt.bar(
    sales_by_location.index,
    sales_by_location.values
)

plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Location Type')

plt.xticks(rotation=45)

plt.show()
```
